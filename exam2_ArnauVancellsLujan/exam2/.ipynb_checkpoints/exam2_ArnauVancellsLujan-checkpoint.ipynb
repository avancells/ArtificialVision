**1.1) (0.5 points)** Read images `obamaImage.jpg` and `Noisymaskbatman.jpg` and display them.
<img src="imagestext/obama1.png" width="500" height="1000">


```python
import numpy as np # useful for manipulating matrices
from skimage import io # useful for image loading and saving
from matplotlib import pyplot as plt # useful for plotting images on iPython or Jupyter Notebooks
```

```python
obama = io.imread('imagesExamen/obamaImage.jpg') #importing the images from the folders
mask = io.imread('imagesExamen/Noisymaskbatman.jpg')

# i define a function to print images, as in this exam we will need to print images
# many times with the same methodology, so we don't need to "copy-paste" this code again
def print_1x2(imatge,imatge_despres, title1, title2):
    fig, axarr = plt.subplots(1,2)
    fig.set_size_inches(10,10)
    
    #imatge
    axarr[0].imshow(imatge,cmap='gray')
    axarr[0].set_title(title1)
    axarr[0].xaxis.set_visible(False)
    axarr[0].yaxis.set_visible(False)
    
    #imatge desprs
    axarr[1].imshow(imatge_despres, cmap='gray')
    axarr[1].set_title(title2)
    axarr[1].xaxis.set_visible(False)
    axarr[1].yaxis.set_visible(False)
    plt.show()
print_1x2(obama,mask, "Obama", "Given Mask")#calling the printing function, with its respective titles
```

**1.2) (0.5 points)** Convert both images in gray images of the same size and display them.

```python
from skimage.color import rgb2gray#importing rgb2gray from skimage.color library
obama_grayscale = obama.copy()#we copy both images to avoid pointer-related memory problems
mask_grayscale = mask.copy()

obama_grayscale = rgb2gray(obama)#we use the rgb2gray function to convert the images
mask_grayscale = rgb2gray(mask)

print_1x2( obama_grayscale, mask_grayscale, "Obama Grayscale", "Mask Grayscale")
```

**1.3) (1 points)** Smooth the mask image to eliminate the noise.  Visualize the filtered and the smoothed masks.
<img src="imagestext/mask1.png" width="600" height="120">


```python
#in order to smooth the mask, we can use a gaussian filter
from skimage.filters import gaussian #we import the gaussian filter from skimage.filters library
mask_gaussian = mask_grayscale.copy()#we copy both images to avoid pointer-related memory problems
mask_gaussian = gaussian(mask_gaussian, sigma=2)#we apply the filter to our image with sigma = 2
print_1x2(mask_grayscale,mask_gaussian, "Original Mask", "Mask Filtered")
```

**1.3) (2 points)** Binarize the mask to separate it from the background. Use the histogram in order to choose the right threshold. Visualize the filtered and binary masks and the histogram of the filtered mask.
<img src="imagestext/mask2.png" width="600" height="120">


```python
from skimage.color import rgb2gray
from skimage import img_as_float
from skimage import exposure

#this will show us a grayscale histogram
def histograma_1(nbins_in, image):

    hist_gray, bins_gray = exposure.histogram(image, nbins = nbins_in)
    fig, ax = plt.subplots(figsize=(10,2))

    ind = np.arange(nbins_in)
    width = 0.2
    
    
    rects_gray = ax.bar(ind, hist_gray, width, color = 'y')
    hist_value = np.where(hist_gray == max(hist_gray))
    
    plt.show()
#this functin will return the threshold for the mask detection
def histograma_1_max(nbins_in, image):

    hist_gray, bins_gray = exposure.histogram(image, nbins = nbins_in)
    fig, ax = plt.subplots(figsize=(10,2))

    ind = np.arange(nbins_in)
    width = 0.2
    
    
    rects_gray = ax.bar(ind, hist_gray, width, color = 'y')
    hist_value = np.where(hist_gray == max(hist_gray))
    return(255/(16-hist_value[0][0]))

#we save the value of the threshold
threshold = histograma_1_max(16,mask_gaussian)

mask_bool=np.zeros([mask_grayscale.shape[0],mask_grayscale.shape[1]]) 

#We create a boolean array, using the data from the grayscale image, this array will have its value
# set to 1 where a value from the grayscale image is greater than threshold (divided by 255 to scale it)

mask_bool[mask_gaussian >threshold/255]=1 


print_1x2(mask_gaussian, mask_bool, "Mask", "Binary Mask")
```

**1.4) (1 point)** Overlap the mask on the Obama image

<img src="imagestext/obama2.png" width="200" height="50">


```python
from skimage.transform import resize
obama_final = obama_grayscale.copy()
mask_resize = resize(mask_bool, (obama_final.shape[0],obama_final.shape[1]), mode = 'reflect')


for i in range(obama_final.shape[0]):
    for j in range(obama_final.shape[1]):
        if(mask_resize[i,j]==0):
            obama_final[i,j]=0

plt.imshow(obama_final,cmap='gray')
plt.show()
```

**2.2) (1 points)** Segment the Jolie image using the felzenszwalb segmentation and determine which set of parameters is better: 
i) (scale=50, sigma=5, min_size=50), or
ii) (scale=100, sigma=20, min_size=150). Explain the difference.

Visualize the original image, the greyscale image and the result of the segmentation in the same figure (using subplots).

```python
def print_1x3(imatge,imatge1, imatge2, title, title1, title2):
    fig, axarr = plt.subplots(1,3)
    fig.set_size_inches(20,10)
    
    #imatge
    axarr[0].imshow(imatge,cmap='gray')
    axarr[0].set_title(title)
    axarr[0].xaxis.set_visible(False)
    axarr[0].yaxis.set_visible(False)
    
    axarr[1].imshow(imatge1, cmap='gray')
    axarr[1].set_title(title1)
    axarr[1].xaxis.set_visible(False)
    axarr[1].yaxis.set_visible(False)
    
    axarr[2].imshow(imatge2, cmap='gray')
    axarr[2].set_title(title2)
    axarr[2].xaxis.set_visible(False)
    axarr[2].yaxis.set_visible(False)
    plt.show()
```

```python
from skimage.segmentation import felzenszwalb
import skimage

segmentation_image = jolie.copy()
segmentation_image2 = jolie.copy()

segmentation_image = felzenszwalb(segmentation_image, scale = 50, sigma = 5, min_size = 50)
segmentation_image2 = felzenszwalb(segmentation_image2, scale = 100, sigma = 20, min_size = 150)

print_1x3(jolie,segmentation_image,skimage.segmentation.mark_boundaries(jolie, segmentation_image),"Original","Parameters 1","Boundaries 1")
print_1x3(jolie,segmentation_image2,skimage.segmentation.mark_boundaries(jolie, segmentation_image2),
          "Original","Parameters 2","Boundaries 2")


segmentation_image_grayscale = jolie_grayscale.copy()
segmentation_image2_grayscale = jolie_grayscale.copy()

segmentation_image_grayscale = felzenszwalb(segmentation_image_grayscale, scale = 50, sigma = 5, min_size = 50)
segmentation_image2_grayscale = felzenszwalb(segmentation_image2_grayscale, scale = 100, sigma = 20, min_size = 150)

print_1x3(jolie_grayscale,segmentation_image_grayscale,
          skimage.segmentation.mark_boundaries(jolie_grayscale, segmentation_image_grayscale),
          "Original Grayscale","Parameters 1","Boundaries 1")
print_1x3(jolie_grayscale,segmentation_image2_grayscale,
          skimage.segmentation.mark_boundaries(jolie_grayscale, segmentation_image2_grayscale),
          "Original Grayscale","Parameters 2","Boundaries 2")

```

The difference between the parameters 1 and parameters 2, depends on each parameter by itself.
Modifying the minSize we are changing the minimum size of a segment to be "true" or "accepted" for what we are looking for.
Sigma will modify the size of the gaussian filtering.
And the scale will change the size of the clusters, bigger size, bigger clusters.

As the scale on the second parameters and the min size is way bigger, same with the sigma, the felzenswalb segmentation isn't finding enough good segments with this parameters, this parameters are way too big for what we are looking for,we can see that the first parameters result is better than the second ones.

**2.4) (1.5 points)** Apply the Canny detector on the Jolie in order to emphasize the borders of the mask using sigma=1 and sigma=2. Overlay the obtained edges on the original image so that the edges with sigma=1 should appear in white and with sigma=2 should appear red as follows.

<img src="./imagestext/jolie.png"  width="500" height="130">

```python
from skimage import feature
jolie_canny1 = jolie.copy()
jolie_canny2 = jolie.copy()
jolie_canny1 = feature.canny(rgb2gray(jolie_canny1), sigma = 1)
jolie_canny2 = feature.canny(rgb2gray(jolie_canny2), sigma = 2)
jolie_sigma_1 = jolie.copy()
jolie_sigma_2 = jolie.copy()


for i in range(jolie_sigma_1.shape[0]):
    for j in range(jolie_sigma_1.shape[1]):
        if(jolie_canny1[i,j]==True):
            jolie_sigma_1[i,j]=[255,255,255]


for i in range(jolie_sigma_2.shape[0]):
    for j in range(jolie_sigma_2.shape[1]):
        if(jolie_canny2[i,j]==True):
            jolie_sigma_2[i,j]=[255,0,0]
            

print_1x3(jolie_grayscale,jolie_sigma_1,jolie_sigma_2,"Grayscale","Sigma 1", "Sigma 2")
print_1x3(jolie_grayscale,jolie_canny1,jolie_canny2,"Grayscale","Sigma 1", "Sigma 2")
```
