###Â 1.1 Image creation

Remember that [Numpy](https://docs.scipy.org/doc/numpy-1.13.0/reference/) allows us to define images of different kinds treating them as matrices.

The most common examples are:

* Binary: with values of 0 or 1, interpreted as black or white, respectively.
* Grayscale: with possible data types uint8, uint16, double, whose pixel values define their intensity.
* RGB Image: color images consisting of 3 matrices (usually called 'channels'), which contain intensity values for each color separately R (red), G (green) and B (blue).

###Â 1.5 Contrast enhancement

a) Open the image `car_lowContrast.jpg` and visualize it without graylevel enhancement. To this purpose, check which are the minimal and maximal values and how to use the command imshow without contrast enhancement.

```python
cotxe = io.imread('images/car_lowContrast.jpg')
plt.imshow(cotxe, cmap= 'gray')
plt.show()
print(cotxe.min())
print(cotxe.max())
```

b) Create an image called `dark_image` using the  values from the original one, so that the minimal value of the new image is 0. Check the minimal and maximal values of the new image, and compare them with the previous one.

```python
cotxe2 = io.imread('images/car_lowContrast.jpg')
dark_image = cotxe2.copy()
dark_image = cotxe2-cotxe2.min()#we substract the minimum value to every pixel from the image
plt.imshow(dark_image, cmap = 'gray')
plt.show()
print(dark_image.min())
print(dark_image.max())
```

c) Create a new image called `bright_im` from the original one, so that the maximal value of the new image is 255. Check the minimal and maximal values of the new image, and compare them with the previous one.

```python
cotxe3 = io.imread('images/car_lowContrast.jpg')
bright_im = cotxe3.copy()
bright_im = cotxe3+255-cotxe3.max()#we sum the difference between our max value and 255 to every pixel from the image
plt.imshow(bright_im, cmap = 'gray')
plt.show()
print(bright_im.min())
print(bright_im.max())
```

d) Apply contrast enhancement to the original image. There are two ways for it: (1) by direct manipulation, and (2) through a scikit-image command. Check the minimal and maximal values of the enhanced image, and compare them with the previous one.

```python
cotxe4 = io.imread('images/car_lowContrast.jpg')
minim = cotxe4.min()
maxim = cotxe4.max()
for i in range(cotxe4.shape[0]):
    for j in range(cotxe.shape[1]):
        cotxe4[i,j] = (cotxe4[i,j]-minim)/(maxim-minim)*255 #we apply the standardization formula to 
                                                            #set the range of values from the image from 0 to 255


plt.imshow(cotxe4, cmap = 'gray')
plt.show()
print(cotxe4.min())
print(cotxe4.max())
```

```python
from skimage import exposure
cotxe5 = io.imread('images/car_lowContrast.jpg')
cotxe5 = exposure.rescale_intensity(cotxe5)#using rescale_intensity to apply contrast enhancement
plt.imshow(cotxe5, cmap = 'gray')
plt.show()
print(cotxe5.min())
print(cotxe5.max())
```

### 1.7 Color manipulation

a) Create and visualize the three images shown in the following figure:
<img src="./images_notebook/ex17a.png">

```python
#using different conditions we set the values of this images to 255 or 0
imageA = np.ndarray([100,100], np.uint8)
for i in range(imageA.shape[0]):
    for j in range(imageA.shape[1]):
        if (j<50):
            imageA[i,j]=0
        else:
            imageA[i,j]=255
            
imageB = np.ndarray([100,100], np.uint8)
for i in range(imageB.shape[0]):
    for j in range(imageB.shape[1]):
        if (i<50):
            imageB[i,j]=0
        else:
            imageB[i,j]=255
            
imageC = np.ndarray([100,100], np.uint8)
for i in range(imageC.shape[0]):
    for j in range(imageC.shape[1]):
        if (i>50 or j>50):
            imageC[i,j]=0
        else:
            imageC[i,j]=255

            
imatges = [imageA, imageB, imageC]
titols = ["", "", ""]

for i in range(3):
    plt.subplot(2,2,i+1),plt.imshow(imatges[i], cmap = 'gray')
    plt.title(titols[i])
    plt.xticks([]),plt.yticks([])
plt.show()            
```

b) Using these three images, construct the following color image:
<img src='./images_notebook/ex17b.png'>

```python
imageFinal = np.ndarray([100,100,3],np.uint8)

#using the images that we previously created, we set the colors for a new image
#this image is using the imageA as the Red channel
#the imageB as the Green channel
#the imageC as the Blue channel
for i in range(imageFinal.shape[0]):
    for j in range(imageFinal.shape[1]):
        imageFinal[i,j] = [imageA[i,j],imageB[i,j],imageC[i,j]]


plt.imshow(imageFinal, cmap='gray')
plt.show()
```

d) Construct and visualize a new image in which the red color of the chairs is changed to blue (see figure).
<img src="images_notebook/ex17c3.png">

```python
coixins = io.imread('images/sillas.jpg')
#To visualize a new image where the red color from the pillows is changed to blue we will 
#swap the Red Channel with the Blue channel, so everything we see as Red from the pillows will
#be seen as blue, while the other pixels of the image will most likely look the same
#the image may look a bit more blue-ish than the original image
for i in range(coixins.shape[0]):
    for j in range(coixins.shape[1]):
        temp =  coixins[i,j][0]
        coixins[i,j][0] = coixins[i,j][2]
        coixins[i,j][2] = temp
        

plt.imshow(coixins, cmap = 'gray')
plt.show()
```

e) Given the images `coat.png` and `model.png` (first two images on the figure below), implement the necessary code to construct the image of the model with the coat (third image). Next find in internet an image with a texture and apply it to obtain an image of the jacket with the found texture (fourth image is an example).

<img src="images_notebook/ex17d.png">

**Note:** when applying resize, rotate, etc. keep track how the image type and values range change. When necessary, apply type conversion using img_as_float, img_as_uint, img_as_ubyte.

```python
from skimage.color import rgb2gray
vestit = io.imread('images/coat.png')
model = io.imread('images/model.png')

#we save a grayscale image of the coat
vestit_grayscale = rgb2gray(vestit)

#we create an array of 0s with the shape of the original image
vestit_bool=np.zeros([vestit_grayscale.shape[0],vestit_grayscale.shape[1]]) 
#we use this new array to set to true the values of the image we want to copy
#so we will be copying everything from the coat image that has color on it (that it is different than 0)
vestit_bool[vestit_grayscale !=0]=1

#then we change the model image values, setting the coat pixels that are different to 0 to the model image
for i in range(vestit_bool.shape[0]):
    for j in range(vestit_bool.shape[1]):
        if vestit_bool[i,j] == 1:
            model[i,j] = vestit[i,j]

plt.imshow(model, cmap = 'gray')
plt.show()

```

```python
from skimage.color import rgb2gray
vestit = io.imread('images/coat.png')
model2 = io.imread('images/model.png')
textura = io.imread('images/textura-de-papel.jpg')

#we save a grayscale image of the coat
vestit_grayscale = rgb2gray(vestit)
#we create an array of 0s with the shape of the original image
vestit_bool=np.zeros([vestit_grayscale.shape[0],vestit_grayscale.shape[1]])
#we use this new array to set to true the values of the image we want to copy
#so we will be copying everything from the coat image that has color on it (that it is different than 0)
vestit_bool[vestit_grayscale!=0]=1

#then we change the model image values, setting the coat pixels that are different to 0 to the model image
#but now we are copying the data from the texture image instead of the coat, considering that the image has transparency
#we have to add the value of transparency to the texture values we are copying(we use the values from the model image)
for i in range(vestit_bool.shape[0]):
    for j in range(vestit_bool.shape[1]):
        if vestit_bool[i,j] == 1:
            model2[i,j] = np.append(textura[i,j], model2[i,j][3])
            
plt.imshow(model2, cmap = 'gray')
plt.show()
```
